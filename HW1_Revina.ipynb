### Самое важное про пробелы

**Не использовать** пробелы: 
* внутри круглых, фигурных, квадратных скобок 
* перед запятой, точкой с запятой и двоеточием 
* перед открывающей скобкой, после которой начинается список агрументров при вызове функции/индекс/срез 

Перед операторами присваивания, сравнения и логическими **нужно** использовать пробелы. Но! вокруг знака равно, если он используется для обозначения именнованного аргумента или значения параметра, ставить пробелы *не стоит*. 

### Самое важное про комментарии 
Итак, *комментарии не должны противоречить коду*. Ни-ког-да. 
* Исправляете код -- исправляете комментарий. 
* Комментарий -- законченное предложение. После точки -- два пробела. 
* Каждая строчка комментария начинается с #, после # -- пробел 
* Блок комментариев должен объяснять весь код. Обычно его пишут в начале.  
* Встрочные комментарии, т.е. находящиеся на одной строчке с кодом, лучше не использовать, это отвлекает от чтения

### Самое важное про именования
Главное: *имена, которые видны пользователю как часть общественного API должны следовать конвенциям, которые отражают использование, а не реализацию.*

**Стили имен**:
* b - одиночная маленькая
* B - одиночная заглавная 
* lowercase - нижний регистр 
* lower_case - нижний регистр + подчеркивания 
* UPPERCASE - заглавные буквы 
* UPPER_CASE - заглавные буквы + подчеркивания 
* CapWords
* mixedCase
* Mixed_Case

**Символы, которые нельзя использовать**: 
* l/I - непонятно, 1 или L, I  
* O - непонятно, это ноль или "о" 

**Общие характеристики имен**: 
* Краткие, нижний регистр 
* Подчеркивания -- хорошо, улучшают читаемость

Важно: не менять регистр переменной, иначе -- это другая переменная. Но создавать две переменные, отличающиеся по стилю вряд ли нужно.
